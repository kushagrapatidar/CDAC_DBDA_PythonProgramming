```python
# Activity 1
# String Operations
strng1='String'
strng2='Operations'
print(strng1, strng2+':', end='\n\n')

concat_strng=strng1+strng2
print('Concatnated string: ', concat_strng, end='\n\n')

repeat_strng=strng1*3
print('Repeated string: ', repeat_strng, end='\n\n')

member_strng='ing' in concat_strng
print('"ing" in concatenated string: ', member_strng, end='\n\n')

sliced_strng=concat_strng[3:14]
print('Sliced string from 3-14 ', sliced_strng)
```

```python
# Activity 2
# String Methods
strng1='Hello'
strng2='World'

print('Capitalize:', strng1.capitalize())
print('Upper:', strng1.upper())
print('Lower:', strng1.lower())
print('Swapcase:', strng1.swapcase())
print('Title:', strng1.title())
print('Find:',strng1.find('ing'))
print('Replace:', strng1.replace('l', 'y'))

strng='Hello, world'
strng1,strng2=strng.split(',')
print('Split:', strng1, strng2)
strng2=strng2.strip()
print('Strip:', strng2)
print(strng2)
print('Join:', '-'.join((strng1,strng2)))
```

```python
# Activity 3
# List Methods, Operations and Functions
lst=[12,14,1,16,15,13,24]
print('Original list:', lst)

lst.append(25)
print('Appended list:', lst)

lst.insert(1,0)
print('Inserted list:', lst)

lst.remove(1)
print('Removed list:', lst)

lst.pop()
print('Popped list:', lst)

lst2=[12,14,1,16,19,20,89]
print('List2 :', lst2)

concat_lst=lst+lst2
print('Concatenated list:', concat_lst)

repeat_lst=[0]*3
print('Repeated list:', repeat_lst)

member_lst=15 in concat_lst
print('15 in concatenated list:', member_lst)

length_lst=len(lst)
print('Length of list:', length_lst)

sliced_lst=concat_lst[3:11]
print('Sliced list from 3-11:', sliced_lst)

max_lst=max(concat_lst)
print('Maximum element:', max_lst)

min_lst=min(concat_lst)
print('Minimum element:', min_lst)

index=concat_lst.index(max_lst)
print('Index of maximum element:', index)

count=concat_lst.count(12)
print('Count of 12:', count)

sorted_lst=sorted(concat_lst)
print('Sorted list:', sorted_lst)

reversed_lst=concat_lst[::-1]
print('Reversed list:', reversed_lst)

concat_lst_copy=concat_lst.copy()
print('Copied list:', concat_lst_copy)

concat_lst.clear()
print('Cleared list:', concat_lst)

del lst2

```

```python
# Activity 4
# Tuple Methods and Functions
tpl=(12,14,1,16)
tpl2=(15,13,24)

concat_tpl=tpl+tpl2
print('Concatenated tuple:', concat_tpl)

sliced_tpl=concat_tpl[1:6]
print("Sliced tuple", sliced_tpl)

repeat_tpl=tuple([0]*4)
print('Repeated tuple:', repeat_tpl)

member_tpl=14 in concat_tpl
print('14 in concatenated tuple:', member_tpl)

length_tpl=len(concat_tpl)
print('Length of tuple:', length_tpl)

max_tpl=max(concat_tpl)
print('Maximum element:', max_tpl)

min_tpl=min(concat_tpl)
print('Minimum element:', min_tpl)

index=concat_tpl.index(max_tpl)
print('Index of maximum element:', index)

del tpl
```

```python
# Activity 5
# Dictionary Methods and Functions
dict1={'name':'John', 'age':25, 'city':'New York'}
dict2={'occupation':'Engineer', 'country':'USA'}


dict2.get('occupation','Chartered Accountant')
print(dict2)

print(dict1.pop('name'))
print(dict1)

print(dict2.popitem())
print(dict2)

print(len(dict1))

print(str(dict1))

print(type(dict1))

print(dict1.copy())

print(dict1.fromkeys('age'))

print(dict1.keys())

print(dict1.values())

dict2.clear()
print(dict2)
```

```python
# Activity 6
# Set Operations, Functions and Methods
set1={12,14,1,16,15,13,24}
set2={12,14,1,16,19,20,89}

union_set=set1.union(set2)
print('Union:', union_set)

print('Intersection:', set1.intersection(set2))

print('Difference:', set1.difference(set2))

print('Symmetric Difference:', set1.symmetric_difference(set2))

print(len(union_set))

print(max(union_set))

print(min(union_set))

print(sorted(union_set))

union_set.add(55)
print(union_set)

union_set_copy=union_set.copy()
print(union_set_copy)

union_set.discard(12)
print(union_set)

union_set.remove(16)
print(union_set)

union_set.pop()
print(union_set)

union_set.clear()
print(union_set)

print(set1)
print(set2)

print("Symmetric Difference Update:",set1.symmetric_difference_update(set2))
```
